```
from newspaper import Article 
import nltk
nltk.download('punkt')
  
def tokenize_article(url):
  the_article = Article(url, language="en") # en for English 
  #To download the article 
  the_article.download() 
  #To parse the article 
  the_article.parse() 
  #To perform natural language processing ie..nlp 
  the_article.nlp() 
  #To extract title 
  print("Article's Title:") 
  print(the_article.title) 
  #To extract text 
  print("Article's Text:") 
  print(the_article.text) 
  #To extract summary 
  print("Article's Summary:") 
  print(the_article.summary)  
```

```
import requests
from bs4 import BeautifulSoup

page = requests.get('http://www.moneycontrol.com') 
soup = BeautifulSoup(page.content, 'html.parser')
for link in soup.find_all('a'):
    temp_link=link.get('href')
    #print(type(temp_link))
    if(temp_link!=None and temp_link.endswith(".html")):
      print(temp_link)
      tokenize_article(temp_link)

```
